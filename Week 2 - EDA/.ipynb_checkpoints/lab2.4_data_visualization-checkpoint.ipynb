## Cleaning and Visualization of market customer data

In this notebook, after importing the dataset we will analyze it through the following visualizations:

- Total number of responses count_plot
- Response by sales channel count_plot
- Response by total claim amount count_plot (we will divide the claim amount in percentiles to search for any additional insight)
- Response by income count_plot (same as before)
- Response by region count_plot

At the end we will take a look at possible relationship between the data with an heatmap and investigate deeper the ones we find

```python
# First we import the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("marketing_customer_analysis_3.csv")
df
```

```python
# Total number of responses count_plot

ax = sns.countplot(x=df['response'])
ax.set_title('Total number of responses')
abs_values = df['response'].value_counts().values
#abs_values = df['response'].value_counts().values
#ax.bar_label(container=ax.containers[0], labels=abs_values)
ax.bar_label(container=ax.containers[0])
```

```python
# Response by sales channel count_plot

ax = sns.countplot(x=df['sales_channel'],hue=df['response'])
ax.set_title('Number of response by sales channel')
ax.bar_label(container=ax.containers[0])
ax.bar_label(container=ax.containers[1])
```

```python
# Total claim amount values distribution
claim_amount_df = df[['total_claim_amount']].sort_values('total_claim_amount')

y = np.array(claim_amount_df['total_claim_amount'])
x = np.array(range(len(y)))
plt.plot(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
plt.title("Claim amount values distribution")
```

```python
# Response by total claim amount count_plot (we will divide the claim amount in percentiles to search for any additional insight)

percentiles = {
    '1-20%': df['total_claim_amount'].quantile(0.2),
    '21-40%': df['total_claim_amount'].quantile(0.4),
    '41-60%': df['total_claim_amount'].quantile(0.6),
    '61-80%': df['total_claim_amount'].quantile(0.8),
    '81-100%': df['total_claim_amount'].quantile(1.0),
}

df['claim_percentiles'] = None
for key in percentiles:
    value = percentiles[key]
    for i in df.index:
        if not df.loc[i,'claim_percentiles'] and df.loc[i,'total_claim_amount'] < value:
            df.loc[i,'claim_percentiles'] = key


ax = sns.countplot(x=df['claim_percentiles'],hue=df['response'],order=df['claim_percentiles'].value_counts().index.sort_values())
ax.set_title('Number of response by claim amount percentiles')
ax.bar_label(container=ax.containers[0])
ax.bar_label(container=ax.containers[1])
```

```python
# Response by income count_plot (same technique as previous plot)

percentiles = {
    '1-20%': df['income'].quantile(0.2),
    '21-40%': df['income'].quantile(0.4),
    '41-60%': df['income'].quantile(0.6),
    '61-80%': df['income'].quantile(0.8),
    '81-100%': df['income'].quantile(1.0),
}

df['income_percentiles'] = None
for key in percentiles:
    value = percentiles[key]
    for i in df.index:
        if not df.loc[i,'income_percentiles'] and df.loc[i,'income'] < value:
            df.loc[i,'income_percentiles'] = key
            

ax = sns.countplot(x=df['income_percentiles'],hue=df['response'],order=df['income_percentiles'].value_counts().index.sort_values())
ax.set_title('Number of response by income percentiles')
ax.bar_label(container=ax.containers[0])
ax.bar_label(container=ax.containers[1])
```

```python
# Response by region count_plot

ax = sns.countplot(x=df['region'],hue=df['response'])
ax.set_title('Number of response by region')
ax.bar_label(container=ax.containers[0])
ax.bar_label(container=ax.containers[1])
```

```python
# Heatmap considering also the response: only one correlation found 

fig = plt.figure(figsize=(8,5))
df['response'] = df['response'].apply(lambda x: 1 if x=='yes' else 0)
numerical = df.select_dtypes(include=np.number)
corr = numerical.corr()
sns.heatmap(corr,annot=True,mask=np.triu(corr))
```

```python
# Visualizing the correlation between monthly_premium_auto and total_claim_amount
# with scatter plot and linear regression analysis

x = np.array(numerical['monthly_premium_auto'])
y = np.array(numerical['total_claim_amount'])

plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
```
