1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("marketing_customer_analysis.csv", index_col=0)
df1 = df[(df['Total Claim Amount']>1000) & (df['Response']=='Yes')]
df1
```

2. Using the original dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
df_yes = df[df['Response']=='Yes']
df_yes.groupby(['Policy Type','Gender'])[['Total Claim Amount']].mean()
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
cust_df = df.groupby(['State'])[['Customer']].count()
cust_df
```

```python
cust_df.loc[cust_df['Customer']>2000]
```

```python
df.groupby(['State']).filter(lambda x: x['Customer'].count()>2000).groupby(['State'])[['Customer']].count()
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
df.groupby(['Education','Gender'])['Customer Lifetime Value'].agg(['max','min','median']).round(2)
```

5. The marketing team wants to analyze the number of policies sold by state and month.

Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
df
```

```python
#df['State'].fillna(df['State'].mode()[0], inplace=True)
```

```python
def jan(s):
    s = s[s < '2/01/11']
    return len(s)

def feb(s):
    s = s[s >= '2/1/11']
    return len(s)

df1 = df.groupby('State').agg({'Effective To Date': [jan,feb]})
df1
#df1.to_numpy().sum()
```

6.  Display a new DataFrame that contains the number of policies sold by month by state for the top 3 states with the highest number of policies sold.

*Hint:* 
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
df1 = df.groupby('State').agg({'Effective To Date': [jan,feb]})
df1.reset_index(inplace=True)
df1.columns = ['State', 'jan','feb']
df1['tot'] = df1['jan'] + df1['feb']
df1.sort_values('tot', ascending=False, inplace=True)
df1 = df1.head(3)
df1
```

7. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

```python
def yes_rate(s):
    return round(len(s[s == 'Yes'])/len(s)*100,2)

df.groupby('Sales Channel')[['Response']].agg(yes_rate)
```
