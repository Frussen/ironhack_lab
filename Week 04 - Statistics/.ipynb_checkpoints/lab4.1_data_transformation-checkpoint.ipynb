```python
# Use the file Data_Marketing_Customer_Analysis_Round3.csv
# Check if there are highly correlated features and drop them (if there are any).

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OrdinalEncoder

df = pd.read_csv('Data_Marketing_Customer_Analysis_Round3.csv')

numerical = df.select_dtypes(include=np.number)
categorical = df.select_dtypes(include=object)

corr_df = numerical.corr()
sns.heatmap(corr_df,annot=True,mask=np.triu(corr_df))
```

```python
# One Hot/Label Encoding of the categorical variables in the categoricals data frame that you created in Activity 3. ( Hint: check Dropfirst option).

def ord_encoding(x):
    dict_encoding = {
        'jan': 1,
        'feb': 2,
        'small': 1,
        'medsize': 2,
        'large': 3,
        'high school or below': 1,
        'college': 2,
        'bachelor': 3,
        'master': 4,
        'doctor': 5,
    }
    x = x.apply(lambda y: dict_encoding[y])
    return x

ord_encoded = categorical[['education','month','vehicle_size']]
ord_encoded = ord_encoded.apply(ord_encoding)
ord_encoded
```

```python
hot_encoded = categorical.drop(['education','month','vehicle_size'], axis=1)
hot_encoded = pd.get_dummies(hot_encoded, drop_first=True)
hot_encoded
```

```python
cat_encoded = pd.concat([ord_encoded,hot_encoded], axis=1)
cat_encoded
```

```python
# Begin by visually examining distributions (histograms) of the numerical features.

import warnings
warnings.filterwarnings('ignore', category=UserWarning)

n_rows=2
n_cols=4
# Create the subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,5))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)

for i, column in enumerate(numerical.columns):
    sns.histplot(numerical[column], ax=axes[i//n_cols,i%n_cols], kde = True)
    #sns.distplot(numerical[column], ax=axes[i//n_cols,i%n_cols])
```

```python
# Select a variable, call it varA, which takes on a wide range of numerical values, and another, varB, which has a noticeably large skew. For example, you might select customer_lifetime_value as a candidate varB which has skew.

varA = (numerical.max() - numerical.min()).idxmax()    # income
varB = (numerical.skew()).idxmax()                     # customer_lifetime_value
print('customer_lifetime_value skew: ', numerical[varB].skew())

# Use minmax transform to bring varA's values into the range [0,1].
minmax = MinMaxScaler()
minmax.fit(np.array(numerical[varA]).reshape(-1,1))
numerical[varA] = minmax.transform(np.array(numerical[varA]).reshape(-1,1))

# Check that varA has been rescaled using a displot or a histogram
n_rows=2
n_cols=4
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,5))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i, column in enumerate(numerical.columns):
    sns.histplot(numerical[column], ax=axes[i//n_cols,i%n_cols], kde = True)
```

```python
# Use StandardScaler to reduce the skew and standardize varB's distribution.
num1 = numerical.copy()
scaler = StandardScaler()
scaler.fit(np.array(num1[varB]).reshape(-1,1))
num1[varB] = scaler.transform(np.array(num1[varB]).reshape(-1,1))
print('customer_lifetime_value skew: ', num1[varB].skew())
print('customer_lifetime_value mean: ', num1[varB].mean())
print('customer_lifetime_value unit variance: ', num1[varB].std())

# Check that the result has zero mean, unit variance, and reduced skew using mean(), std(), and a plot of the PDF.
n_rows=2
n_cols=4
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,5))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i, column in enumerate(num1.columns):
    sns.histplot(num1[column], ax=axes[i//n_cols,i%n_cols], kde = True)
```

```python
# Use PowerTransform  with yeo-johnson to reduce the skew and standardize varB's distribution.
num2 = numerical.copy()
transformer = PowerTransformer(standardize=False)
transformer.fit(np.array(num2[varB]).reshape(-1,1))
num2[varB] = transformer.transform(np.array(num2[varB]).reshape(-1,1))
print('customer_lifetime_value skew: ', num2[varB].skew())
print('customer_lifetime_value mean: ', num2[varB].mean())
print('customer_lifetime_value unit variance: ', num2[varB].std())

# Check that the result has zero mean, unit variance, and reduced skew using mean(), std(), and a plot of the PDF.
n_rows=2
n_cols=4
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,5))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i, column in enumerate(num2.columns):
    sns.histplot(num2[column], ax=axes[i//n_cols,i%n_cols], kde = True)
```

```python
# Use PowerTransform  with box-cox to reduce the skew and standardize varB's distribution.
#yeo-johnson
num3 = numerical.copy()
transformer = PowerTransformer(method='box-cox', standardize=False)
transformer.fit(np.array(num3[varB]).reshape(-1,1))
num3[varB] = transformer.transform(np.array(num3[varB]).reshape(-1,1))
print('customer_lifetime_value skew: ', num3[varB].skew())
print('customer_lifetime_value mean: ', num3[varB].mean())
print('customer_lifetime_value unit variance: ', num3[varB].std())

# Check that the result has zero mean, unit variance, and reduced skew using mean(), std(), and a plot of the PDF.
n_rows=2
n_cols=4
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10,5))
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i, column in enumerate(num3.columns):
    sns.histplot(num3[column], ax=axes[i//n_cols,i%n_cols], kde = True)
```
